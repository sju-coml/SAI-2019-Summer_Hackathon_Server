# Normallize and PCA

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

test set 분리

Normallize and PCA

```python
# Data Preperation (exclude all categorical data)
df_dropped_data_num = df_dropped_data.iloc[:,1:np.size(df_dropped_data.iloc[0,:])]

# Normalization
scaler = StandardScaler()
scaler.fit(df_dropped_data_num)
df_dropped_data_num_scaled = scaler.transform(df_dropped_data_num)
```

```python
dropped_data_num = PCA(n_components=np.size(df_dropped_data.iloc[0,1:]))
dropped_data_num_res = dropped_data_num.fit(df_dropped_data_num_scaled)
dropped_data_num
```

```python
column_list = []
for i in range(1, np.size(df_dropped_data.iloc[0,1:])+1, 1):
    column_list.append("pc_" + str(i))
```

```python
# Important Results of PCA
PCA_result = ['Variance', 'Variance\%','Cum\%']
dropped_data_num_res_pc = np.transpose(np.array(dropped_data_num_res.components_).reshape(np.shape(dropped_data_num_res.components_)))
dropped_data_num_res_var = np.array(np.transpose(dropped_data_num_res.explained_variance_[:,np.newaxis]))
dropped_data_num_res_varp = np.array(np.transpose(dropped_data_num_res.explained_variance_ratio_[:,np.newaxis]))
dropped_data_num_res_sum = np.cumsum(np.round(dropped_data_num_res.explained_variance_ratio_, decimals=8)*100)
dropped_data_num_res_varc = np.array(np.transpose(dropped_data_num_res_sum[:,np.newaxis]))
dropped_data_num_res_tab = DataFrame(np.concatenate((dropped_data_num_res_pc,
                                                        dropped_data_num_res_var,
                                                        dropped_data_num_res_varp,
                                                        dropped_data_num_res_varc),
                                                       axis = 0),
                                        index = np.concatenate((np.array(df_dropped_data_num.columns), PCA_result)),
                                        columns = column_list)
dropped_data_num_res_tab
```
